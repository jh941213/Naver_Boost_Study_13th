- 다음의 결과가 나오도록 합니다. 이 때 그래프의 색상, 스타일 표현방식에 차이가 있더라도 상관계수값이 차이가 없다면 문제 없습니다.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# 관심있는 컬럼들만 선택
selected_columns = ["음주여부", "흡연상태", "연령대코드(5세단위)", "성별코드"]
subset = df[selected_columns]

# 상관계수 구하기
correlation = subset.corr()

# 상관계수 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=.5, fmt=".2f")
plt.title("미남홀란드의 상관계수 구하기")
plt.show()
````

### 🌜 절반만 마스킹해서 상관계수 표시하기

````python
# 관심있는 컬럼들만 선택
selected_columns = ["음주여부", "흡연상태", "연령대코드(5세단위)", "성별코드"]
subset = df[selected_columns]

# 상관계수 구하기
correlation = subset.corr()

# numpy를 활용한 마스크
mask = np.triu(np.ones_like(correlation, dtype=bool))

# 상관계수 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=.5, fmt=".2f", mask=mask)
plt.title("미남홀란드의 상관계수 구하기")
plt.show()
````

### 📊 상관계수 히트맵 작성 파라미터 (정리)

- **데이터프레임**: 상관계수를 계산할 데이터를 포함한 데이터프레임을 지정합니다.

- **상관계수 메서드**: 📈 Pearson 상관계수, 🧮 Spearman 상관계수 등 다양한 상관계수 메서드 중 하나를 선택합니다.

- **컬러맵**: 상관계수 값에 따라 색상을 나타내는데 사용되는 컬러맵을 지정합니다.

- **상관계수 값 범위**: 컬러맵에서 사용되는 색상의 범위를 지정합니다. 일반적으로 -1부터 1까지 사용됩니다.

- **히트맵 크기**: 히트맵의 크기를 설정합니다. 그림 크기나 행렬 크기에 따라 조절할 수 있습니다.

- **히트맵 제목**: 히트맵에 제목을 추가할 수 있습니다. 📌 그래프의 내용을 요약하는데 도움이 됩니다.

- **x축 레이블, y축 레이블**: 📋 x축과 y축에 표시되는 레이블을 지정합니다.

- **히트맵 스타일**: 🎨 히트맵의 스타일을 지정할 수 있으며, 배경색, 글꼴 크기 등을 조절할 수 있습니다.

- **색상바**: 상관계수 값을 색상으로 표시하는 색상바를 추가할 수 있습니다.

- **상관계수 값 소수점 자리수**: 상관계수 값의 소수점 자리수를 조절하여 히트맵에 표시되는 값의 정확도를 설정합니다.

- **상관계수 값 표시 여부**: 히트맵에 상관계수 값 자체를 표시할지 여부를 지정합니다.

- **상관계수 값에 대한 텍스트 색상**: 상관계수 값의 긍정적인 또는 부정적인 연관성을 표시하기 위한 글자 색상을 지정합니다.

- **추가적인 플롯 설정**: 히트맵을 사용할 시 추가적인 플롯 설정을 적용할 수 있습니다. 예를 들어, 여러 개의 서브플롯을 생성하는 등의 설정을 추가할 수 있습니다.

````python
smoke = {1 : "흡연안함", 2: "끊음", 3: "흡연중"}
drink = {0: "안마심", 1: "마심"}

# 데이터프레임의 값 변경
df["흡연상태"] = df["흡연상태"].map(smoke)
df["음주여부"] = df["음주여부"].map(drink)

# 교차 테이블 생성
cross_tab = pd.crosstab(df["흡연상태"], df["음주여부"])
````

#### **🌝 FacetGrid를 사용하여 그래프 그리기**

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip", encoding="cp949")

# 매핑 딕셔너리
smoke = {1 : "흡연안함", 2: "끊음", 3: "흡연중"}
drink = {0: "안마심", 1: "마심"}

# 체중이 120Kg 이상인 데이터 필터링
over_120kg = df[df['체중(5Kg 단위)'] >= 120]

# 매핑 딕셔너리를 이용하여 데이터프레임의 값 변경
over_120kg["흡연상태"] = over_120kg["흡연상태"].map(smoke)
over_120kg["음주여부"] = over_120kg["음주여부"].map(drink)

# FacetGrid를 사용하여 그래프 그리기
# sns.regplot 을 추가하면 회귀선이 기록된다 tip
g = sns.FacetGrid(over_120kg, col="흡연상태", hue="음주여부")
g.map(sns.regplot, "총콜레스테롤", "감마지티피", scatter_kws={'s':10}, ci=None)

# 그래프 제목과 축 레이블 추가
g.set_axis_labels("총콜레스테롤", "감마지티피")

g.add_legend(title="음주여부")
plt.show()
````

````python
age_code = {1: '0~4세',
 2: '5~9세',
 3: '10~14세',
 4: '15~19세',
 5: '20~24세',
 6: '25~29세',
 7: '30~34세',
 8: '35~39세',
 9: '40~44세',
 10: '45~49세',
 11: '50~54세',
 12: '55~59세',
 13: '60~64세',
 14: '65~69세',
 15: '70~74세',
 16: '75~79세',
 17: '80~84세',
 18: '85세+'}
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip", encoding="cp949")

# 실명 데이터 제거
df_filtered = df[df['시력(좌)'] != 9.9]
df_filtered = df_filtered[df_filtered['시력(우)'] != 9.9]

df_filtered['연령대코드(5세단위)'] = df_filtered['연령대코드(5세단위)'].map(age_code)

# 연령대와 성별에 따른 좌, 우 평균 시력 계산
grouped = df_filtered.groupby(['연령대코드(5세단위)', '성별코드']).agg({'시력(좌)': 'mean', '시력(우)': 'mean'}).reset_index()

# 서브플롯 생성
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# 좌 시력
sns.barplot(data=grouped, x='시력(좌)', y='연령대코드(5세단위)', hue='성별코드', ax=ax[0])
ax[0].set_title("연령대 및 성별에 따른 좌 시력 평균")
ax[0].set_xlabel("평균 좌 시력")
ax[0].legend(title="성별")

# 우 시력
sns.barplot(data=grouped, x='시력(우)', y='연령대코드(5세단위)', hue='성별코드', ax=ax[1])
ax[1].set_title("연령대 및 성별에 따른 우 시력 평균")
ax[1].set_xlabel("평균 우 시력")
ax[1].legend(title="성별")

plt.tight_layout()
plt.show()
````
